```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import re
```

```python
links = []
for page in range(1, 31):
    url = 'https://weartesters.com/category/performance-reviews/page/{}/'.format(page)
    res = requests.get(url)
    con = res.content
    soup = BeautifulSoup(con, 'lxml')
    for articles in soup.find('main'):
        if articles.find('h2') == -1:
            pass
        else:
            reviews = articles.find('h2').find('a').attrs['href']
            links.append(reviews)
```

```python
links[0:5]
```

```python
shoes = []
for shoe in links:
    sneaks = shoe.replace('https://weartesters.com/', '').replace('nightwing230', '').replace('update', '').replace('zak kerr/', '').replace('jahronmon', '').replace('thewongkicks/', '').replace('gets update/', '').replace('tsestanley2', '').replace('-performance-review/', '').replace('anotherpair_/', '').replace('-', ' ').replace('nyjumpman2', '').replace('3/', '').replace('theyre better original/', '').replace(' duke4005/', '').replace('re /', '').replace('performance', '').replace('review', '').replace(' performance review', '').replace(' quick kicks/', '').replace(' tsestanley23/', '').replace(' nyjumpman23/', '').replace(' bryanhinkle1/', '').replace(' nightwing2303/', '').replace(' stanley t/', '').replace(' re review/', '').replace(' jahronmon/', '').replace(' anotherpair/', '').replace(' 2/','')
    shoes.append(sneaks)
```

```python
shoes = {"shoes" : shoes}
```

```python
shoes = pd.DataFrame(shoes)
```

```python
wts = []
for link in links:
    url = '{}'.format(link)
    reqs = requests.get(url)
    cont = reqs.content
    bsoup = BeautifulSoup(cont, 'lxml')
    reviews = bsoup.find('main').find_all('p')
    for review in reviews:
        wts.append(review.get_text())
```
